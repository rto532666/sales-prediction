## We have train (8523) and test (5681) data set, train data set has both input and output



### Attribute Information

* Item_Identifier: Unique product ID
* Item_Weight: Weight of product
* Item_Fat_Content: Whether the product is low fat or not
* Item_Visibility: The % of total display area of all products in a store allocated to the
* particular product
* Item_Type: The category to which the product belongs
* Item_MRP: Maximum Retail Price (list price) of the product
* Outlet_Identifier: Unique store ID
* Outlet_Establishment_Year: The year in which store was established
* Outlet_Size: The size of the store in terms of ground area covered
* Outlet_Location_Type: The type of city in which the store is located
* Outlet_Type: Whether the outlet is just a grocery store or some sort of supermarket
* Item_Outlet_Sales: Sales of the product in the particulat store. This is the outcome
* variable to be predicted.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("../dataset/Train.csv")
sales
```

```python
sales.shape
```

```python
# Checking null values in feature
sales.isnull().sum()
```

```python
# Replacing nan in Item_Weight with its mean (feature contains continious value)
sales['Item_Weight'].fillna(sales.Item_Weight.mean(),inplace=True)
```

```python
sales.Outlet_Size.value_counts()
```

```python
# Replacing nan in Outlet_Size with its mode (feature contains categorical value)
sales['Outlet_Size']=sales['Outlet_Size'].fillna(sales.Outlet_Size.value_counts().index[0])
```

```python
sales.isnull().sum()
```

```python
# Handling categorical feature
df_cat=sales[['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type']]
```

```python
df_cat.head()
```

```python
# Checking the length of unique values in feature
for feature in df_cat:
    print(feature,':',len(df_cat[feature].unique()))
```

```python
sales.Item_Fat_Content.unique()
```

```python
sales['Item_Fat_Content']=np.where(sales['Item_Fat_Content']=='low fat','Low Fat',sales['Item_Fat_Content'])
sales['Item_Fat_Content']=np.where(sales['Item_Fat_Content']=='LF','Low Fat',sales['Item_Fat_Content'])
sales['Item_Fat_Content']=np.where(sales['Item_Fat_Content']=='reg','Regular',sales['Item_Fat_Content'])
sales.Item_Fat_Content.unique()
```

```python
sales['Item_Fat_Content']=pd.get_dummies(sales['Item_Fat_Content'],drop_first=True)
sales.head()
```

```python
# Checking the unique values

print(sales.Item_Fat_Content.unique())
print(sales.Item_Type.unique())
print(sales.Outlet_Size.unique())
print(sales.Outlet_Location_Type.unique())
print(sales.Outlet_Type.unique())
```

```python
sales.drop(['Item_Identifier','Outlet_Identifier'],axis=1,inplace=True)
sales.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating a function to implement LabelEncoder

encoding=LabelEncoder()
def encoder(sales,feature):
    sales[feature]=encoding.fit_transform(sales[feature])
```

```python
# Extracting categorical feature name and passing it to function.

for columns in df_cat:
    feature=columns
    encoder(sales,feature)
```

```python
# Checking the unique values

print(sales.Item_Fat_Content.unique())
print(sales.Item_Type.unique())
print(sales.Outlet_Size.unique())
print(sales.Outlet_Location_Type.unique())
print(sales.Outlet_Type.unique())
```
